### Data Cleaning & Missing Values

```python
## Creating data with missing values for data cleaning demonstration

# dict
dirty_data = {
    'name': ['Alice', 'Bob', None, 'Diana', 'Eve'],
    'age': [25, None, 35, 28, 32],
    'salary': [50000, 60000, 70000, None, 65000],
    'email': ['alice@email.com', 'bob@email', 'charlie@email.com', 
              'diana@email.com', None]
}

dirty_df = pd.DataFrame(dirty_data)

dirty_df.head()
```

```python
## Detecting missing values
dirty_df.isnull()  # Returns a boolean mask | if particular cell is None then return True for that cell
```

```python
dirty_df.isnull().sum()  # Count missing per column
```

```python
dirty_df.info()  # Shows not null counts
```

```python
## Handling missing values

# Drop rows with any missing values
clean_df = dirty_df.dropna()  # dropna() will drop rows if the rows have any missing values

clean_df.head()
```

```python
# Drop rows with missing values in specific columns
clean_df = dirty_df.dropna(subset=['name', 'age'])

clean_df
```

```python
# Fill missing values

filled_df = dirty_df.fillna({
    'age': dirty_df['age'].mean(),
    'salary': dirty_df['salary'].median(),
    'name': 'Unknown'
})

filled_df

# Note: need more example on this.
```

### Merging & Joining Data

```python
## Sample data for merging
employees = pd.DataFrame({
    'emp_id': [1, 2, 3, 4],
    'name': ['Alice', 'Bob', 'Charlie', 'Diana'],
    'dept_id': [1, 2, 1, 3]
})

departments = pd.DataFrame({
    'dept_id': [1, 2, 3, 4],
    'dept_name': ['IT', 'HR', 'Finance', 'Marketing']
})
```

```python
## Inner Join (only matching records)

inner_merged = pd.merge(employees, departments, on='dept_id', how='inner')
inner_merged
```

```python
## Left join (all records from left table)
left_merged = pd.merge(employees, departments, on='dept_id', how='left')
left_merged
```

```python
## Right join (all records from right table)
right_merged = pd.merge(employees, departments, on='dept_id', how='right')
right_merged
```

```python
## Outer join (all records from both tables)
outer_merged = pd.merge(employees, departments, on='dept_id', how='outer')

outer_merged
```

```python
df1 = pd.DataFrame({
    'key1': [1, 2, 3],
    'value': ['A', 'B', 'C']
})

df1
```

```python
df2 = pd.DataFrame({
    'key2': [1, 2, 4],
    'value': ['X', 'Y', 'Z']
})

df2
```

```python
## Merging on different column names

# when we want to merge the 2 dataframe but the keys are not the same

merged = pd.merge(df1, df2, left_on='key1', right_on='key2', how='outer')

merged
```

```python
## Concatenating DataFrames

# Vertical concatenation (stacking rows)
df_concat_rows = pd.concat([employees, departments], ignore_index=True)

df_concat_rows

# as we can see that there is differnt keys so concatination perform that way
```

```python
# Now if we perform concatination with same keys let check the output now
df_concat_rows = pd.concat([employees, employees], ignore_index=True)

df_concat_rows
```

```python
# Horizontal concatenation (side by side)
df_concat_cols = pd.concat([employees, departments], axis=1)
df_concat_cols
```
